```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    
class LinkedList():
    def __init__(self):
        self.head = None
    
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        
    def display(self):
        curr = self.head
        items = []
        while curr:
            items.append(str(curr.data))
            curr = curr.next
        print(" -> ".join(items))
    


ll = LinkedList()
ll.append(1)
ll.display()
ll.append(1)
ll.display()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
    
class LinkedList:
    def __init__(self):
        self.head = None
    
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        last = self.head
        while last.next:
            last = last.next
        last.next = new_node
        
    def display(self):
        items = []
        curr = self.head
        while curr:
            items.append(str(curr.data))
            curr = curr.next
        print(" -> ".join(items))
        
    def insert_start(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_after(self, prev_node, data):
        curr = self.head
        while curr and curr.data != prev_node:
            curr = curr.next
        if not curr:
            print("Given node not in list")
            return
        new_node = Node(data)
        new_node.next = curr.next
        curr.next = new_node
        
linkedlist = LinkedList()
linkedlist.append(1)
linkedlist.append(2)
linkedlist.append(4)
linkedlist.append(5)
linkedlist.append(6)
linkedlist.display()
linkedlist.insert_start(0)
linkedlist.display()
linkedlist.insert_after(2, 3)    
linkedlist.display()
```
